```python
import csv
import html
import io
import json
import sys
import zipfile
from itertools import cycle
from pathlib import Path
```

```python
# 1. Load every CSV from the zip into a dict {filename: list[dict]}

def load_zip(zip_path: str) -> dict[str, list[dict[str, str]]]:
    data: dict[str, list[dict[str, str]]] = {}
    with zipfile.ZipFile(zip_path) as zf:
        for name in zf.namelist():
            if name.lower().endswith(".csv"):
                key = Path(name).name
                raw = zf.read(name).decode("utf-8-sig")
                reader = csv.DictReader(io.StringIO(raw))
                data[key] = [dict(row) for row in reader]
    return data
```

```python
# 2. Parse helpers

def text(val) -> str:
    return "" if val is None else str(val).strip()


def num(val) -> float:
    """Strip currency/percent symbols and return float."""
    cleaned = (
        text(val)
        .replace("\u20b9", "")
        .replace("Ã¢â€šÂ¹", "")
        .replace(",", "")
        .replace("%", "")
    )
    if not cleaned:
        return 0.0
    try:
        return float(cleaned)
    except ValueError:
        return 0.0


def fmt_inr(n: float) -> str:
    return f"\u20b9{n:,.0f}"


def safe_upper(val) -> str:
    return text(val).upper()


def trunc(val: str, limit: int) -> str:
    val = text(val)
    return val if len(val) <= limit else val[:limit] + "..."
```

```python
# 3. Extract structured data from each CSV

def parse(data: dict[str, list[dict[str, str]]]) -> dict:
    out: dict = {}

    fi = {
        text(r.get("Particulars")): text(r.get("Details"))
        for r in data.get("01_firm_info.csv", [])
        if text(r.get("Particulars"))
    }
    out["firm"] = fi

    out["partners"] = [
        {
            "name": text(r.get("Partner Name")),
            "ratio": num(r.get("Profit Sharing Ratio (%)")),
        }
        for r in data.get("02_partners.csv", [])
        if text(r.get("Partner Name"))
    ]

    out["liabilities"] = {
        text(r.get("Item")): num(r.get("Amount (INR)"))
        for r in data.get("03_balance_sheet_liabilities.csv", [])
        if text(r.get("Item"))
    }

    out["assets_bs"] = {
        text(r.get("Item")): num(r.get("Amount (INR)"))
        for r in data.get("04_balance_sheet_assets.csv", [])
        if text(r.get("Item"))
    }

    debtors = [
        {
            "name": text(r.get("Debtor Name")),
            "amount": num(r.get("Amount (INR)")),
        }
        for r in data.get("05_sundry_debtors.csv", [])
        if text(r.get("Debtor Name")) and safe_upper(r.get("Debtor Name")) != "TOTAL"
    ]
    out["debtors"] = sorted(debtors, key=lambda x: x["amount"], reverse=True)

    creditors = [
        {
            "name": text(r.get("Creditor Name")),
            "amount": num(r.get("Amount (INR)")),
        }
        for r in data.get("06_sundry_creditors.csv", [])
        if text(r.get("Creditor Name")) and safe_upper(r.get("Creditor Name")) != "TOTAL"
    ]
    out["creditors"] = sorted(creditors, key=lambda x: x["amount"], reverse=True)

    fixed = [
        {
            "name": text(r.get("Asset Group")),
            "rate": text(r.get("Depreciation Rate (%)")),
            "opening": num(r.get("WDV 01-04-2024 (INR)")),
            "additions": num(r.get("Additions (INR)")),
            "dep": num(r.get("Depreciation (INR)")),
            "closing": num(r.get("WDV 31-03-2025 (INR)")),
        }
        for r in data.get("07_fixed_assets.csv", [])
        if text(r.get("Asset Group")) and safe_upper(r.get("Asset Group")) != "TOTAL"
    ]
    out["fixed_assets"] = fixed

    trading: dict[str, list[dict[str, float | str]]] = {}
    for r in data.get("08_trading_account.csv", []):
        side = text(r.get("Side"))
        item = text(r.get("Particulars"))
        if side and item:
            trading.setdefault(side, []).append({"item": item, "amount": num(r.get("Amount (INR)"))})
    out["trading"] = trading

    pl: dict[str, list[dict[str, float | str]]] = {}
    for r in data.get("09_profit_and_loss.csv", []):
        side = text(r.get("Side"))
        item = text(r.get("Particulars"))
        if side and item:
            pl.setdefault(side, []).append({"item": item, "amount": num(r.get("Amount (INR)"))})
    out["pl"] = pl

    cap: dict[str, list[dict[str, float | str]]] = {}
    for r in data.get("10_partners_capital.csv", []):
        partner = text(r.get("Partner"))
        item = text(r.get("Particulars"))
        if partner and item:
            cap.setdefault(partner, []).append({"item": item, "amount": num(r.get("Amount (INR)"))})
    out["capital"] = cap

    ratios = {
        text(r.get("Parameter")): {
            "cy": text(r.get("Current Year 2024-25 (INR)")),
            "py": text(r.get("Previous Year 2023-24 (INR)")),
        }
        for r in data.get("11_key_ratios.csv", [])
        if text(r.get("Parameter"))
    }
    out["ratios"] = ratios

    out["observations"] = [
        {"no": text(r.get("S.No.")), "text": text(r.get("Observation"))}
        for r in data.get("12_tax_audit_observations.csv", [])
        if text(r.get("Observation"))
    ]

    return out
```

```python
# 4. Derive chart-ready values

def derive(p: dict) -> dict:
    d: dict = {}

    ratios = p.get("ratios", {})
    d["turnover_cy"] = num(ratios.get("Total Turnover (Sales)", {}).get("cy", "0"))
    d["turnover_py"] = num(ratios.get("Total Turnover (Sales)", {}).get("py", "0"))
    d["gp_cy"] = num(ratios.get("Gross Profit", {}).get("cy", "0"))
    d["gp_py"] = num(ratios.get("Gross Profit", {}).get("py", "0"))
    d["gp_ratio_cy"] = num(ratios.get("Gross Profit Ratio", {}).get("cy", "0"))
    d["gp_ratio_py"] = num(ratios.get("Gross Profit Ratio", {}).get("py", "0"))
    d["np_cy"] = num(ratios.get("Net Profit", {}).get("cy", "0"))
    d["np_py"] = num(ratios.get("Net Profit", {}).get("py", "0"))
    d["np_ratio_cy"] = num(ratios.get("Net Profit Ratio", {}).get("cy", "0"))
    d["np_ratio_py"] = num(ratios.get("Net Profit Ratio", {}).get("py", "0"))
    d["closing_stock"] = num(ratios.get("Closing Stock", {}).get("cy", "0"))
    d["stock_pct"] = num(ratios.get("Closing Stock / Turnover", {}).get("cy", "0"))

    d["turnover_yoy"] = (
        (d["turnover_cy"] - d["turnover_py"]) / d["turnover_py"] * 100
        if d["turnover_py"]
        else 0
    )

    d["total_debtors"] = sum(x["amount"] for x in p.get("debtors", []))
    d["total_creditors"] = sum(x["amount"] for x in p.get("creditors", []))

    bs = p.get("assets_bs", {})
    d["stock_5"] = bs.get("Closing Stock 5% GST", 0)
    d["stock_18"] = bs.get("Closing Stock 18% GST", 0)
    d["stock_12"] = bs.get("Closing Stock 12% GST", 0)
    d["stock_zero"] = bs.get("Closing Stock Zero Rated GST", 0)

    dr = {i["item"]: i["amount"] for i in p.get("trading", {}).get("DR", [])}
    d["open_stock_5"] = dr.get("Opening Stock (5%) GST", 0)
    d["open_stock_18"] = dr.get("Opening Stock (18%) GST", 0)
    d["open_stock_12"] = dr.get("Opening Stock (12%) GST", 0)
    d["open_stock_taxfree"] = dr.get("Opening Stock (Tax Free)", 0)

    d["purch_5"] = dr.get("Purchase GST (5%)", 0)
    d["purch_18"] = dr.get("Purchase GST (18%)", 0)
    d["purch_12"] = dr.get("Purchase GST (12%)", 0)
    d["purch_taxfree"] = dr.get("Purchase GST (Tax Free)", 0)
    d["purch_28"] = dr.get("Purchase GST (28%)", 0)

    cr = {i["item"]: i["amount"] for i in p.get("trading", {}).get("CR", [])}
    d["sales_18"] = cr.get("Sales GST (18%)", 0)
    d["sales_12"] = cr.get("Sales GST (12%)", 0)
    d["sales_5"] = cr.get("Sales GST (5%)", 0)
    d["sales_taxfree"] = cr.get("Sales GST (Tax Free)", 0)

    pl_dr = {i["item"]: i["amount"] for i in p.get("pl", {}).get("DR", [])}
    pl_cr = {i["item"]: i["amount"] for i in p.get("pl", {}).get("CR", [])}
    d["expenses"] = sorted(
        [(k, v) for k, v in pl_dr.items() if k != "Net Profit (transferred)"],
        key=lambda x: x[1],
        reverse=True,
    )
    d["total_income"] = sum(i["amount"] for i in p.get("pl", {}).get("CR", []))
    d["total_expenses"] = sum(
        i["amount"]
        for i in p.get("pl", {}).get("DR", [])
        if i["item"] != "Net Profit (transferred)"
    )
    d["gp_from_trading"] = pl_cr.get("Gross Profit (from Trading)", 0)
    d["discount_received"] = pl_cr.get("Discount Received", 0)
    d["net_profit"] = pl_dr.get("Net Profit (transferred)", d["np_cy"])

    d["cap_movements"] = {}
    for partner, rows in p.get("capital", {}).items():
        d["cap_movements"][partner] = {r["item"]: r["amount"] for r in rows}

    return d
```
